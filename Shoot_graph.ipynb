```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("shooting1.txt") as f1:
    data1 = f1.readlines()
with open("shooting2.txt") as f2:
    data2 = f2.readlines()
with open("shooting3.txt") as f3:
    data3 = f3.readlines()
with open("shooting4.txt") as f4:
    data3 = f4.readlines()
```

```python
x1 = []
x2 = []
x3 = []
x4 = []

y1 = []
y2 = []
y3 = []
y4 = []

for line in data1:
    x1.append(float(line[:line.find('\t')]))
    y1.append(float(line[line.find('\t'):]))
for line in data2:
    x2.append(float(line[:line.find('\t')]))
    y2.append(float(line[line.find('\t'):]))
for line in data3:
    x3.append(float(line[:line.find('\t')]))
    y3.append(float(line[line.find('\t'):]))
for line in data4:
    x4.append(float(line[:line.find('\t')]))
    y4.append(float(line[line.find('\t'):]))
```

```python
def u0(x):
    return np.exp(-np.sinh(x)) + x
```

```python
x = np.linspace(0, 1, 50)
y = [u0(t) for t in x]
```

```python
plt.figure(figsize=(15, 12))
plt.plot(x, y, 'g', linewidth = 6, label = 'Exact solution')

plt.plot(x1[0], y1[0], 'ro', label = 'Step: 0.01')
plt.plot(x1[0:2], y1[0:2], 'r')
for i in range(2, len(x1)):
    plt.plot(x1[i-1], y1[i-1], 'ro')
    plt.plot(x1[i-1:i+1], y1[i-1:i+1], 'r')
    
plt.plot(x2[0], y2[0], 'bo', label = 'Step: 0.02')
plt.plot(x2[0:2], y2[0:2], 'b')
for i in range(2, len(x2)):
    plt.plot(x2[i-1], y2[i-1], 'bo')
    plt.plot(x2[i-1:i+1], y2[i-1:i+1], 'b')

plt.plot(x3[0], y3[0], 'yo', label = 'Step: 0.05')
plt.plot(x3[0:2], y3[0:2], 'y')
for i in range(2, len(x3)):
    plt.plot(x3[i-1], y3[i-1], 'yo')
    plt.plot(x3[i-1:i+1], y3[i-1:i+1], 'y')
#
#plt.plot(x4[0], y4[0], 'co', label = 'Step: 0.1')
#plt.plot(x4[0:2], y4[0:2], 'c')
#for i in range(2, len(x4)):
#    plt.plot(x4[i-1], y4[i-1], 'co')
#    plt.plot(x4[i-1:i+1], y4[i-1:i+1], 'c')

plt.legend(fontsize = 20)
plt.title("Shooting method", fontsize = 24)
plt.grid()
plt.show()
```

```python
err1 = max([abs(y1[i] - u0(x1[i])) for i in range(len(y1))])
err2 = max([abs(y2[i] - u0(x2[i])) for i in range(len(y2))])
err3 = max([abs(y3[i] - u0(x3[i])) for i in range(len(y3))])
err4 = max([abs(y4[i] - u0(x4[i])) for i in range(len(y4))])
```

```python
lgerr = [np.log(err1), np.log(err2), np.log(err3), np.log(err4)]
h = [0.01, 0.02, 0.05, 0.1]
lgh = [np.log(x) for x in h]
```

```python
plt.figure(figsize=(15, 12))
plt.plot(lgh, lgerr, '--')
plt.title("Error", fontsize = 24)
plt.xlabel("lg(h)", fontsize = 24)
plt.ylabel("lg(err)", fontsize = 24)
plt.grid()
```
